# Analysing the dataset and creating graphs using matplotlib and seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('one.csv.csv')
df
```

```python
print(df.head())
```

```python
#Histogram
sns.histplot(df['release_year'],kde=True)
plt.xlabel('Release year')
plt.ylabel('Frequency');
plt.title("Histogram")
plt.show()
```

```python
#scatter plot
sns.scatterplot(x='release_year',y='duration',data=df)
plt.xlabel('Release year')
plt.ylabel('Duration')
plt.show()
```

```python
#visualising length of the movie
df_movie= df[df['type']=='Movie']
plt.hist(df_movie['duration'])
```

```python
#checking if the average rating have been changed overtime
df['rating'].value_counts().plot(kind='line')
df.groupby('release_year')['rating'].value_counts().unstack().plot()
df.groupby('listed_in')['rating'].value_counts().unstack(0).plot(kind='line')
```
