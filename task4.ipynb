# Train a simple linear regression model on dataset and predict the output

# 

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

#train data
tr_data= pd.read_csv('train.csv')

#dropping null rows in column y
tr_data = tr_data.dropna(subset=['y'])
x_train = tr_data[['x']].astype(float)
y_train = tr_data[['y']].astype(float)


#linear regression model for train
model=LinearRegression()
model.fit(x_train,y_train)

#test data
te_data = pd.read_csv('Test.csv')
x_test=te_data[['x']].astype(float)

#predicting the test data
y_pred = model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
x_test=te_data[['x']].astype(float)
y_test=te_data[['y']].astype(float)

#linear regression model for train
model=LinearRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

y_test=y_test.squeeze()
y_pred=y_pred.squeeze()

results=pd.DataFrame({'Actual':y_test,' predicted':y_pred})
print(results)

mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)

metrics=pd.DataFrame({'MSE':[mse],'R2':[r2]})
print(metrics)
```
